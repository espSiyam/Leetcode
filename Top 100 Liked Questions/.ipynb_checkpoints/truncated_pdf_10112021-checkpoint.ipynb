```python
import numpy as np; import matplotlib.pyplot as plt; import scipy.integrate as spi
```

## define function to calculate area under the curve

```python
def get_approx(c, d):
    f = lambda x : (1/(0.00128*0.88726))*((0.00128/(x+0.000132))**2.4)*(np.exp(-0.00128/(x+0.000132))) #full PDF
    # x and y values for the trapezoid rule
    a = -5; b = 10; N = 200
    x1 = np.log10(c); x2 = np.log10(d)
    x = np.logspace(x1,x2,N+1); y = f(x)
    approximation = spi.trapz(y,x)
    
    # plot function, X and Y values for plotting y=f(x)
    X = np.logspace(a,b,100); Y = f(X)
    
    return approximation, X, Y, x, y

def plot_fill(x, **kwargs):
    f = lambda x : (1/(0.00128*0.88726))*((0.00128/(x+0.000132))**2.4)*(np.exp(-0.00128/(x+0.000132))) #full PDF
    N=200; y = f(x)
    color  = kwargs.pop('color'); edgecolor  = kwargs.pop('edgecolor')
    for i in range(N):
        xs = np.array(([x[i],x[i],x[i+1],x[i+1]]))
        ys = [0,f(x[i]),f(x[i+1]),0]
        plt.fill(xs,ys, color=color, edgecolor=edgecolor, alpha = 0.1)
```

## get area under the curve for different bins

```python
# bins (area in km2) = (0.00001 - 10), (0.00001 - 0.0001), (0.0001 - 0.0025), (0.0025 - 0.01), (0.01 - 0.04), (0.04 - 10)
approx, X, Y, x, y = get_approx(0.00001, 10)
approx1, X1, Y1, x1, y1 = get_approx(0.00001, 0.0001)
approx2, X2, Y2, x2, y2 = get_approx(0.0001, 0.0025)
approx3, X3, Y3, x3, y3 = get_approx(0.0025, 0.01)
approx4, X4, Y4, x4, y4 = get_approx(0.01, 0.04)
approx5, X5, Y5, x5, y5 = get_approx(0.04, 10)

#probability is area under the curve for each bin
prob = np.array((approx1, approx2, approx3, approx4, approx5))
prob.round(3)
```

## plot figure

```python
plt.figure(figsize=(7, 7))
plt.plot(X,Y, color='grey')
plt.plot(X1,Y1, color='grey', label = 'P(LS_bin1) ={:5.0f}'.format((approx1*100).round(1)) + '%')
plt.plot(X2,Y2, color='grey', label = 'P(LS_bin2) ={:5.0f}'.format((approx2*100).round(1)) + '%')
plt.plot(X3,Y3, color='grey', label = 'P(LS_bin3) ={:5.0f}'.format((approx3*100).round(1)) + '%')
plt.plot(X4,Y4, color='grey', label = 'P(LS_bin4) ={:5.0f}'.format((approx4*100).round(1)) + '%')
plt.plot(X5,Y5, color='grey', label = 'P(LS_bin5) ={:5.0f}'.format((approx5*100).round(1)) + '%')

plot_fill(x, **{'color': 'k'}, **{'edgecolor': 'k'})
plot_fill(x1, **{'color': 'k'}, **{'edgecolor': 'k'})
plot_fill(x2, **{'color': 'grey'}, **{'edgecolor': 'grey'})
plot_fill(x3, **{'color': 'dimgrey'}, **{'edgecolor': 'dimgrey'})
plot_fill(x4, **{'color': 'silver'}, **{'edgecolor': 'silver'})
plot_fill(x5, **{'color': 'white'}, **{'edgecolor': 'white'})
# plt.title('Trapezoid Rule, N = {}'.format(N))
plt.xlim(1e-5,10); plt.ylim(1e-5,1e3)
plt.xscale('log'); plt.yscale('log')
plt.xlabel("Landslide Area ($km^2$)",fontsize=12) 
plt.ylabel("Probability Density ($km^{-2}$)",fontsize=12) 
plt.legend(handlelength=0)
plt.show()
```

## Now looking to truncate distribution to area at 200 m2 or 0.0002 km2

```python
plt.figure(figsize=(7, 7))
f = lambda x : (1/(0.00128*0.88726))*((0.00128/(x+0.000132))**2.4)*(np.exp(-0.00128/(x+0.000132))) #full PDF
g = lambda z : (1/(0.00128*0.88726))*((0.00128/(z+0.000132))**2.4)*(np.exp(-0.00128/(z+0.000132))) #truncated PDF
# x and y values for the trapezoid rule
a = np.log10(0.00001); b = np.log10(10); c = np.log10(200/1e6); N = 200
x1 = a; x2 = b; z1 = a; z2 = c
x = np.logspace(x1,x2,N+1); y = f(x)
z = np.logspace(z1,z2,N+1); y1 = g(z)

approximation = spi.trapz(y,x)
approximation_trunc = spi.trapz(y1,z)
print(approximation.round(3)); print(approximation_trunc.round(3))
# plot function, X and Y values for plotting y=f(x)
X = np.logspace(a,b,100); Y = f(X)
Z = np.logspace(a,c,100); Y1 = g(Z)
plt.plot(X,Y, label = 'P(LS) = {:5.0f}'.format((approximation*100).round(1)) + '%')
plt.plot(Z,Y1, label = 'P(LS_trunc) = {:5.0f}'.format((approximation_trunc*100).round(1)) + '%')

#     print('Probabilty of landslides = ', approximation.round(3), (approximation*100).round(1) )
for i in range(N):
    xs = [x[i],x[i],x[i+1],x[i+1]]
    ys = [0,f(x[i]),f(x[i+1]),0]
    plt.fill(xs,ys,'b',edgecolor='b',alpha=0.1)
for i in range(N):
    zs = [z[i],z[i],z[i+1],z[i+1]]
    y1s = [0,g(z[i]),g(z[i+1]),0]
    plt.fill(zs,y1s,'r',edgecolor='r',alpha=0.1)

# plt.title('Trapezoid Rule, N = {}'.format(N))
plt.xlim(1e-5,10); plt.ylim(1e-5,1e3)
plt.xscale('log'); plt.yscale('log')
plt.xlabel("Landslide Area ($km^2$)",fontsize=12) 
plt.ylabel("Probability Density ($km^{-2}$)",fontsize=12) 
plt.legend()
plt.show()
```
